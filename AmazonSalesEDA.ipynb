```python
monthly_count = data['Date'].dt.month.value_counts().sort_index()
monthly_count.index
```

```python
sns.countplot(x= data['Date'].dt.month, data=data, palette='Spectral')
plt.show()
```

```python
colors = ['orangered','greenyellow','cornflowerblue','teal']


monthly_counts = data['Date'].dt.month.value_counts().sort_index()


monthly_counts.index = data.groupby(data['Date'].dt.month)['Date'].first().dt.month_name()[monthly_counts.index]

monthly_counts.plot(
    kind='pie', 
    autopct='%0.1f%%', 
    colors=colors,
    startangle=90
)
plt.title('Distribution of Orders by Month')
plt.ylabel('')
plt.show()
```

```python
df['Courier Status'].value_counts()
```

### The majority of orders are shipped, while a few remain unshipped.

```python
df['Category'] = df['Category'].astype(str)
c_d = df['Category']
plt.figure(figsize = (10,6))
plt.hist(c_d,bins = 18, edgecolor ="red")
plt.xticks(rotation = 90)
plt.show()
```

The top-selling category was 'Set', slightly surpassing 'Kurta'.

```python
check_B2B = df['B2B'].value_counts()
plt.pie(check_B2B, labels = check_B2B.index, autopct = '%1.1f%%')
plt.show()
```

Retailers account for 99.3% of all sales, while wholesalers account for only 0.7%.

```python
top10_state = df['ship-state'].value_counts().head(10)
plt.figure(figsize = (20,10))
sns.countplot(data = df[df['ship-state'].isin(top10_state.index)], x= "ship-state", hue="ship-state", palette = 'bright')
plt.xlabel('State')
plt.ylabel('Orders')
plt.title('Distribution of State')
plt.show()
```

Most of the buyers are from Maharashtra State
